```python
import os
import subprocess

video_dir = "data/val_video"
audio_dir = "data/var_audio"
os.makedirs(audio_dir, exist_ok=True)

ffmpeg_path = "ffmpeg"

for file in os.listdir(video_dir):
    if file.endswith(".mp4"):
        mp4_path = os.path.join(video_dir, file)
        base = os.path.splitext(file)[0]
        wav_path = os.path.join(audio_dir, base + ".wav")
        command = [
            ffmpeg_path, "-i", mp4_path,
            "-vn", "-acodec", "pcm_s16le",
            "-ar", "16000", "-ac", "1",
            wav_path
        ]
        print(f"ðŸ”Š ë³€í™˜ ì¤‘: {file}")
        subprocess.run(command)
```
